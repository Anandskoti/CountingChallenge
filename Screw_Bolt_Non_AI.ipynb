# SCREW AND BOLT

```python
import glob
import cv2
path = glob.glob("/Users/anandkoti/Downloads/ScrewAndBolt_20240713/*.jpg")
scrw_bolt = []
for img1 in path:
    n = cv2.imread(img1)
    scrw_bolt.append(n)
```

```python
import matplotlib.pyplot as plt
img=scrw_bolt[15]
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray');
```

```python
blur=cv2.GaussianBlur(gray,(11,11),0)
plt.imshow(blur,cmap='gray')
```

```python
canny=cv2.Canny(blur,30,150,3)
plt.imshow(canny,cmap='gray')
```

```python
dilated=cv2.dilate(canny,(1,1),iterations=2)
plt.imshow(dilated,cmap='gray')
```

```python
(cnt,heirarcy)=cv2.findContours(dilated.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb,cnt,-1,(0,255,0),2)
plt.imshow(rgb)
```

```python
print("screw and bolts in the image",len(cnt))
```
